### 案例

```python
# 生成随机成绩
np.random.seed(42)
scores = pd.Series(np.random.randint(50, 101, 10), index=['学生'+str(i) for i in range(10)])
print(scores)

```

```python
print("平均成绩：", scores.mean())
print("最高分：", scores.max())
print("最低分：", scores.min())
print(scores[scores>scores.mean()])
print("高于平均分学生人数：",np.sum(np.where(scores > scores.mean(), 1, 0)))
print(len(scores[scores>scores.mean()]))
print(scores[scores>scores.mean()].count())
```

# 案例 销售数据分析


```python
sales = pd.Series([120, 135, 145, 160, 155, 170, 180, 175, 190, 200, 210, 220]
                  ,index = pd.date_range('2022-01-01',periods=12, freq='MS' ))
print(sales)
```

```python
#销量最高的月份
print(sales.idxmax())
print()
# 月环比增长率
print(sales.pct_change())
```

```python
# 季度的平均销量
sales.resample("QS").mean()     # resample  重新采样
```

```python
# 连续两个月增长
a = sales.pct_change()
b = a > 0
print(b)
# 滑动窗口
b[b.rolling(3).sum() == 3].keys().tolist()
```

##   案例

```python
# 每小时销售数据分析
np.random.seed(42)
hourly_sales = pd.Series(np.random.randint(0, 100, 24),
                         index=pd.date_range('2025-01-01', periods=24, freq='h'))
print(hourly_sales)
```

```python
# 按天重采样计算销售额
print('日销售额：',*(hourly_sales.resample("D").sum().tolist()))
```

```python
# 每天营业时间和非营业时间的销售额比例
a = sum(hourly_sales[0:8])    # 左闭又开 不包含 8
c =  hourly_sales.iat[23]
print(a)
print(c)
d = a + c
b = sum(hourly_sales[8:23])
print(b)
print("营业时间销售额占比：%.3f"%(b / (d + b)))
print("非营业时间销售额占比：%.3f"%(d / (d+b)))
print(b/d)
```

```python
print(hourly_sales.between_time("8:00", "22:00").sum())   # 筛选时间段内的数据 Series
business_hours_sales = hourly_sales[(hourly_sales.index.hour>=8)
                        & (hourly_sales.index.hour<=22)]
z = hourly_sales.drop(business_hours_sales.index)
print(z.sum())
```

```python
a = hourly_sales.rolling(3).sum()
print(a.idxmax())
print()
# 销售额最高的三个小时
print(hourly_sales.nlargest(3))
# 最小 nsmallest()
```
