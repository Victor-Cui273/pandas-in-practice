```python
data1 = pd.DataFrame(data, index= [1, 2, 3, 4, 5])
input = data1.iloc[:, 0:2]
```

```python
print(data1)
```

```python
dataT = data1.T
dataT
```

```python
data1["总分"] = data1[["数学", "英语", "物理"]].sum(axis=1)
data1["平均分"] = data1[["数学", "英语", "物理"]].mean(axis=1)    # 取多列
```

```python
data1
```

```python
data1[(data1["数学"] > 90) | (data1["英语"] > 85)]
```

# 案例

```python
data = {
    "产品名称": ["A", "B", "C", "D"],
    "单价": [100, 150, 200, 120],
    "销量": [50, 30, 20, 40]
}
df = pd.DataFrame(data)
df
```

```python
s1 = df["单价"]
s2 = df["销量"]
```

```python
df["销售额"] = s1 * s2
df["总销售额"] = df["单价"]*df["销量"]
```

```python
df
```

```python
print("销售额最高：", df["销售额"].max().tolist())
df.nlargest(1, columns=['销售额'])
```

```python
df.sort_values(by="销售额", ascending=False)
```

# 案例 电商用户行为分析

```python
data = {
    "ID":[101, 102, 103, 104, 105],
    "用户名": ["Alice", "Bob", "Charlie", "David", "Eve"],
    "商品类别":["电子产品", "服饰", "电子产品","家居","服饰"],
    "商品单价":[1200, 300, 800, 150, 200],
    "购买数量":[1, 3, 2, 5, 4]
}
df = pd.DataFrame(data)
df
```

```python
# 每位用户消费金额
df["消费总金额"] = df["商品单价"] * df["购买数量"]
df
```

```python
# 消费金额最高的用户
df.nlargest(2, columns="消费总金额")
```

```python
#所有平均消费
print("%.2f"%(df["消费总金额"].mean()))
```

```python
df[df["商品类别"]=="电子产品"]["购买数量"].sum()
```
